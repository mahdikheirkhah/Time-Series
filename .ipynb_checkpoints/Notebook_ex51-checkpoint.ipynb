```python
import pandas as pd
```

1. Create a `Series` named `integer_series` from 1st January 2010 to 31 December 2020. Each date is associated with the number of days since 1st January 2010. It starts with 0.


```python
time_series = pd.date_range('2010-01-01', '2020-12-31', freq='D')
integer_series = pd.Series(range(len(time_series)), index=time_series)
integer_series
```

2. Using Pandas, compute a 7 days moving average. This transformation smooths the time series by removing small fluctuations. **without using for loop**

```python
moving_avg = integer_series.rolling(window=7, min_periods=1).mean()
moving_avg.head(10)
```
